```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(2, 2, figsize=(10,16))  # 1 row, 2 columns

# Plot the histogram for 'Sleep_Hours' on the first subplot
sns.histplot(df['Age'], bins=10, ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Quantity per Age')

occupation_count = df['Occupation'].value_counts().sort_values()
occupation_count.plot(kind='bar',ax=axes[1,0], color='skyblue')
axes[0,1].set_title('Quantity per Occupation')

country_count = df['Country'].value_counts().sort_values()
country_count.plot(kind='bar',ax=axes[0,1], color='skyblue')
axes[1,0].set_title('Quantity per Country')

gender_count = df['Gender'].value_counts()
gender_count.plot(kind='bar',ax=axes[1,1], color='skyblue')
axes[1,1].set_title('Quantity per Gender')
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 8))  # 1 row, 2 columns

# Plot the histogram for 'Sleep_Hours' on the first subplot
sns.histplot(df['Sleep_Hours'], bins=10, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution Sleep_Hours')

# Plot the histogram for 'Stress_Level' on the second subplot
sns.histplot(df['Work_Hours'], bins=10, ax=axes[1], color='skyblue')
axes[1].set_title('Distribution Work_Hours')

# Plot the histogram for 'Stress_Level' on the second subplot
sns.histplot(df['Physical_Activity_Hours'], bins=10, ax=axes[2], color='skyblue')
axes[2].set_title('Distribution Physical_Activity_Hours')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(2, 2, figsize=(8,8))  # 1 row, 2 columns

# Plot the histogram for 'Sleep_Hours' on the first subplot
severity = df['Severity'].value_counts()
severity.plot(kind='bar',ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Count of Severity')
for container in axes[0, 0].containers:
    axes[0, 0].bar_label(container, fmt='%d')

# Plot the histogram for 'Sleep_Hours' on the first subplot
Stress_Level = df['Stress_Level'].value_counts()
Stress_Level.plot(kind='bar',ax=axes[0,1], color='skyblue')
axes[0,1].set_title('Count of Stress_Level')
for container in axes[0, 1].containers:
    axes[0, 1].bar_label(container, fmt='%d')

# Plot the histogram for 'Stress_Level' on the second subplot
menteal_condition = df['Mental_Health_Condition'].value_counts()
menteal_condition.plot(kind='bar',ax=axes[1,0], color='skyblue')
axes[1,0].set_title('Count of Mental_Health_Condition')
for container in axes[1, 0].containers:
    axes[1,0].bar_label(container, fmt='%d')

# Plot the histogram for 'Stress_Level' on the second subplot
Consultation_History = df['Consultation_History'].value_counts()
Consultation_History.plot(kind='bar',ax=axes[1,1], color='skyblue')
axes[1,1].set_title('Count of Consultation_History')
for container in axes[1, 1].containers:
    axes[1, 1].bar_label(container, fmt='%d')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(2, 2, figsize=(12,8))  # 1 row, 2 columns

severity_sleep = df.groupby('Severity')['Sleep_Hours'].mean()
severity_sleep.plot(kind='bar',ax=axes[1,1])
axes[1,1].set_title('Avg. Sleep_Hours by Severity')
for container in axes[1,1].containers:
    axes[1, 1].bar_label(container, fmt='%.2f')

mental_sleep= df.groupby('Mental_Health_Condition')['Sleep_Hours'].mean()
mental_sleep.plot(kind='bar',ax=axes[0,1])
axes[0,1].set_title('Avg. Sleep_Hours by Mental Condition')
for container in axes[0, 1].containers:
    axes[0, 1].bar_label(container, fmt='%.2f')

stress_sleep= df.groupby('Stress_Level')['Sleep_Hours'].mean()
stress_sleep.plot(kind='bar',ax=axes[1,0])
axes[1,0].set_title('Avg. Sleep_Hours by Stress_Level')
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, fmt='%.2f')

consultation_history_sleep= df.groupby('Consultation_History')['Sleep_Hours'].mean()
consultation_history_sleep.plot(kind='bar',ax=axes[0,0])
axes[0,0].set_title('Avg. Sleep_Hours by Consultation_History')
for container in axes[1, 1].containers:
    axes[0,0].bar_label(container, fmt='%.2f')
plt.subplots_adjust(hspace=0.5)
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(2, 2, figsize=(12,10))  # 1 row, 2 columns

severity_work = df.groupby('Severity')['Work_Hours'].mean()
severity_work.plot(kind='bar',ax=axes[0,0])
axes[0,0].set_title('Avg. Work_Hours by Severity')
for container in axes[0, 0].containers:
    axes[0, 0].bar_label(container, fmt='%.2f')

mental_work= df.groupby('Mental_Health_Condition')['Work_Hours'].mean()
mental_work.plot(kind='bar',ax=axes[0,1])
axes[0,1].set_title('Avg. Work_Hours by Mental Condition')
for container in axes[0, 1].containers:
    axes[0, 1].bar_label(container, fmt='%.2f')

stress_work= df.groupby('Stress_Level')['Work_Hours'].mean()
stress_work.plot(kind='bar',ax=axes[1,0])
axes[1,0].set_title('Avg. Work_Hours by Stress_Level')
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, fmt='%.2f')
    
consultation_history_work= df.groupby('Consultation_History')['Work_Hours'].mean()
consultation_history_work.plot(kind='bar',ax=axes[1,1])
axes[1,1].set_title('Avg. Work_Hours by Consultation_History')
for container in axes[1, 1].containers:
    axes[1, 1].bar_label(container, fmt='%.2f')

plt.subplots_adjust(hspace=0.5)
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(2, 2, figsize=(12,12))  # 1 row, 2 columns

severity_physical = df.groupby('Severity')['Physical_Activity_Hours'].mean().sort_values()
severity_physical.plot(kind='bar',ax=axes[0,0])
axes[0,0].set_title('Avg. Phys_Activ_Hours by Severity')
for container in axes[0, 0].containers:
    axes[0, 0].bar_label(container, fmt='%.2f')

mental_physical= df.groupby('Mental_Health_Condition')['Physical_Activity_Hours'].mean().sort_values()
mental_physical.plot(kind='bar',ax=axes[0,1])
axes[0,1].set_title('Avg. Phys_Activ_Hours by Mental Cond')
for container in axes[0, 1].containers:
    axes[0, 1].bar_label(container, fmt='%.2f')

stress_physical= df.groupby('Stress_Level')['Physical_Activity_Hours'].mean().sort_values()
stress_work.plot(kind='bar',ax=axes[1,0])
axes[1,0].set_title('Avg. Phys_Activ_Hours by Stress_Level')
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, fmt='%.2f')

consultation_history_physical= df.groupby('Consultation_History')['Physical_Activity_Hours'].mean().sort_values()
consultation_history_physical.plot(kind='bar',ax=axes[1,1])
axes[1,1].set_title('Avg. Phys_Activ_Hours by Cons. History')
for container in axes[1, 1].containers:
    axes[1, 1].bar_label(container, fmt='%.2f')

plt.subplots_adjust(hspace=0.3)
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(1,3, figsize=(11,4))  # 1 row, 2 columns

occupation_sleep = df.groupby('Occupation')['Sleep_Hours'].mean().sort_values()
occupation_sleep.plot(kind='bar',ax=axes[0])
axes[0].set_title('Avg. Sleep_Hours by Occupation')

occupation_work= df.groupby('Occupation')['Work_Hours'].mean().sort_values()
occupation_work.plot(kind='bar',ax=axes[1])
axes[1].set_title('Avg. Work_Hours by Occupation')

occupation_physical= df.groupby('Occupation')['Physical_Activity_Hours'].mean().sort_values()
occupation_physical.plot(kind='bar',ax=axes[2])
axes[2].set_title('Avg. Phys.Activ. Hours by Occup.')
```

```python
# Create a figure and a set of subplots
fig, axes = plt.subplots(1,3, figsize=(11,6))  # 1 row, 2 columns

country_sleep = df.groupby('Country')['Sleep_Hours'].mean().sort_values()
country_sleep.plot(kind='bar',ax=axes[0])
axes[0].set_title('Avg. Sleep_Hours by Country')

country_work= df.groupby('Country')['Work_Hours'].mean().sort_values()
country_work.plot(kind='bar',ax=axes[1])
axes[1].set_title('Avg. Work_Hours by Country')

country_physical= df.groupby('Country')['Physical_Activity_Hours'].mean().sort_values()
country_physical.plot(kind='bar',ax=axes[2])
axes[2].set_title('Avg. Phys.Activity Hours by Country')
```

```python
stress_level_proportion = round(df.groupby('Country')['Stress_Level'].value_counts(normalize=True).unstack(),2)
ax = stress_level_proportion.plot(kind='bar',stacked=True, figsize=(10, 6))
plt.title('Proportion of Stress Levels by Country')
plt.xlabel('Country')
plt.ylabel('Proportion of Stress Level')
plt.legend(title='Stress Level')

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='center')

plt.tight_layout()
plt.show()
```
